```python
img2gray = cv.cvtColor(img2,cv.COLOR_BGR2GRAY)
print (img2gray.shape)
plt.imshow(img2gray,cmap='gray')
plt.show()
```

```python
mask_inv = cv.bitwise_not(img2gray)
plt.imshow(mask_inv)
plt.show()
```

```python
white_bg=np.full(img2.shape,255,dtype=np.uint8)
print(white_bg.shape)
plt.imshow(white_bg,cmap='gray')
plt.show()
```
